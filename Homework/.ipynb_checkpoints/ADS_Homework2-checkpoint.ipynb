```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import pandas as pd
%matplotlib inline
data = pd.read_csv('h2data.csv', index_col=0)
corr=stats.pearsonr(data.x,data.y)[0]
print('Correlation={0}'.format(corr))
plt.scatter(data.x, data.y)
```

```python
data = pd.read_csv('h2data.csv', index_col=0)
plt.scatter(data.x,data.y)
llm = smf.ols(formula='y ~ x', data = data)
fitted = llm.fit()
print(fitted.summary())
lmy = fitted.predict(data)
plt.plot(data.x, lmy,'r-')
```

```python
residuals = fitted.resid
plt.hist()
```

```python
e<-summary(lm(data$y ~ data$x))$residuals
ecdf(e)
plot.ecdf(e)
hist(e,main = 'Histogram of errors', xlab = 'Frequency', ylab = 'Range of errors', breaks = 15)
m<-round(mean(e))
std<-sqrt(var(e))
xfit<-seq(-(4*std-m),(4*std+m))
yfit<-dnorm(xfit, mean = m, sd = std)
lines(xfit, yfit*5000, col = 'blue', lwd =2)

#The plots are included in the file.
#They do indicate that this distribution is consistent with the classical regression assumptions.
```
